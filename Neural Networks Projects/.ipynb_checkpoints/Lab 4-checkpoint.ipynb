### Dataset and Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Normalization
from sklearn.preprocessing import MinMaxScaler
```

```python
# Generate synthetic wine data (features: acidity, sugar; labels: good/not good)
np.random.seed(42)
acidity = np.random.uniform(3.0, 4.0, 100)
sugar = np.random.uniform(5.0, 15.0, 100)
quality = (acidity * sugar > 30).astype(int)

# Combine features into a dataset
X = np.column_stack((acidity, sugar))
Y = quality.reshape(-1, 1)

# Plot the data
plt.scatter(acidity[quality == 0], sugar[quality == 0],
            label = "Not Good", color = "red")
plt.scatter(acidity[quality == 1], sugar[quality == 1],
            label = "Good", color = "green")
plt.xlabel("Acidity")
plt.ylabel("Sugar Content")
plt.legend()
plt.title("Wine Quality Data")
plt.show()
```

### Data Normalization

```python
# Normalize features using TensorFlow ’s Normalization layer
normalizer = Normalization (axis = -1)
normalizer.adapt(X)
X_normalized = normalizer (X)
print ("Normalized Features:\n", X_normalized [:5])
```

### Building the Neural Network

```python
# Define the neural network
model = Sequential([tf.keras.Input(shape=(2,)), # Input layer for two features
                    Dense(3, activation='sigmoid', name='hidden_layer'), # Hidden layer with three units
                    Dense(1, activation='sigmoid', name='output_layer') # Output layer for binary classification
                   ])

# print model summary
model.summary()
                    
```

### Training the Model

```python
# Compile the model
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    metrics=['accuracy']
)

# Train the model
history = model.fit(X_normalized, Y, epochs=50, batch_size=10, verbose=1)
```

### Visualizing Decision Boundary

```python
# Generate grid points for decision boundary visualization
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.011))

# Predict probabilities for each grid point
grid_points = np.c_[xx.ravel(), yy.ravel()]
grid_points_normalized = normalizer(grid_points)
predictions = model.predict(grid_points_normalized).reshape(xx.shape)

# Plot decision boundary
plt.contourf(xx, yy, predictions, levels=[0, 0.5, 1],
             alpha=0.6, colors=["red", "green"])
plt.scatter(acidity[quality == 0], sugar[quality == 0],
            label="Not Good", color="red")
plt.scatter(acidity[quality == 1], sugar[quality == 1],
            label="Good", color="green")
plt.xlabel("Acidity")
plt.ylabel("Sugar Content")
plt.legend()
plt.title("Decision Boundary for Wine Quality")
plt.show()  
```
